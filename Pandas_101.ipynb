**Pandas 101**

Import your libraries

```python
import pandas as pd
import numpy as np
```

Create a list

```python
snakes = [
    ['small',38,False],
    ['small',62,True],
    ['medium',55,True],
]
```

Create a dataframe from the list

```python
df = pd.DataFrame(snakes, columns=['Snake_size','Length','Poisonous'])
```

Check how our dataframe look like

```python
df
```

Create a dictionary and transform it into a dataframe

```python
snakes_dict = {
    'Snake_size':["small","small","medium"],
    'Length':[38,62,55],
    'Poisinous':[False,True,True]
}
dt = pd.DataFrame(snakes_dict)
```
