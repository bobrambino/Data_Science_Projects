```python
from sklearn.linear_model import LinearRegression
x = dt['petal_width'].to_numpy().reshape((-1, 1))
y = dt['petal_length'].to_numpy()
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

# Draw observation plot
plt.scatter(x, y, color='blue', label='Observations')
plt.plot(x, y_pred, color='red', linewidth=2, label='Linear Regression')

# Adding labels and legend
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Observation Plot of Linear Regression Model')
plt.legend()

# Show the plot
plt.show()
```

```python
# Calculate the errors
errors = y - y_pred

# Create the "Errors over Predictions" plot using matplotlib
plt.scatter(y_pred, errors, label='Errors over Predictions')

# Adding a horizontal line at y=0 for reference
plt.axhline(0, color='red', linestyle='--', label='Zero Error')

# Adding labels and title
plt.xlabel('Predictions')
plt.ylabel('Errors')
plt.title('Errors over Predictions Plot')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
